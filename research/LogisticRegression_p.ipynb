```
import numpy as np
import pandas as pd
from google.colab import files
import io
import random

class LogisticReegression_P(object):
  def link_function_sigmoid(self,y):
    """
    """
    z = np.exp(-np.array(y))
    return 1/(1+z)
  
  def regression(self,data,weight_vector,theta):
    """
    
    """
    x_var=data.iloc[:,:-1]
    y= []
    for i in range(len(data)):
      y.append(np.dot(weight_vector,x_var.iloc[i,:])+theta)
      #print("------y",y)
    return y
    
  def fit(self,data,weight_vector,theta):
    """
    
    """
    y = self.regression(data,weight_vector,theta)
    #print("error y",y)
    sigmoid_y = self.link_function_sigmoid(y)
    print("error y",sigmoid_y)
    return sigmoid_y
  
  def train(self,data):
    """
    
    """
    weight_vector = np.random.normal(-1,1,len(data.columns)-1)
    y_actual=data.iloc[:,-1]
    y_actual =np.array(y_actual)
    y_predicted =[0]*len(y_actual) 
    theta = random.randint(-1,1)
    cost_func =999
    lr =0.01
    new_weight_vector = [0]*len(weight_vector)
    new_theta =  [0]*len(weight_vector)
    
    for i in range(100):
      print("weights :", weight_vector,"theta",theta)
      y_predicted = self.fit(data,weight_vector,theta)
      
     # print("weights new weight",weight_vector,new_weight_vector)
      for w in range(len(weight_vector)):
        x_vector = data.iloc[:,w]
        x_vector = np.array(x_vector)
        
        #print("error", (np.dot(x_vector,(y_predicted-y_actual).T)))
        new_weight_vector[w] = (-1/len(data)) * (np.dot(x_vector,(y_predicted-y_actual)))
        
        new_theta[w] = (-1/len(data)) * sum(y_predicted-y_actual)
        
        weight_vector[w]+=   new_weight_vector[w] * lr
        theta+= np.mean(new_theta)*lr
      cost_func =  ( -1/len(data)) * (np.sum(y_actual * np.log(y_predicted) + (1-y_actual)*np.log(1-y_predicted)))
      
      print("-----------")
      print("cost ", cost_func)
    
    
if __name__=="__main__":
  obj = LogisticReegression_P()
  data = files.upload()
  data = pd.read_csv(io.BytesIO(data["data2.csv"]))
  obj.train(data)
    
  
```
